```python
categorias = ['A', 'B', 'C', 'D', 'E']
valores = [30, 20, 15, 10, 25]

colores_grafica = ['lightblue', 'lightgreen', 'lightyellow', 'lightcoral', 'lightskyblue']
explode = [0.1, 0, 0, 0, 0]
startangle = 0

plt.pie(valores, labels=categorias, colors=colores_grafica, explode=explode, startangle=startangle, shadow=True, autopct='%1.1f%%')

plt.axis('equal')  # Para asegurar una representación circular

plt.show()
```

```python
categorias = ['A', 'B', 'C', 'D', 'E']
valores = [30, 20, 15, 10, 25]

colores_grafica = ['lightblue', 'lightgreen', 'lightyellow', 'lightcoral', 'lightskyblue']
explode = [0.1, 0, 0.5, 0, 0]
startangle = 0

plt.pie(valores, labels=categorias, colors=colores_grafica, explode=explode, startangle=startangle, shadow=True, autopct='%1.1f%%')

plt.legend(title="Categorías", loc="upper center", bbox_to_anchor=(0.1, 0.5))

plt.axis('equal')  # Para asegurar una representación circular

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generamos una lista de valores
x = np.linspace(0, 2*np.pi, 100)
print(np.pi)
print(x)

# Aplicamos la función elemento a elemento
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(x)
```

```python
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title("Sin(x)")

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title("Cos(x)")

plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title("Tan(x)")

plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title("Exp(x)")

# Ajustar los espacios entre los subgráficos
plt.tight_layout()

# Mostrar la figura con los subgráficos
plt.show()
```

# Ejercicio en clase
## Objetivo
Aplicar técnicas avanzadas de visualización de datos para analizar las ventas mensuales de productos electrónicos.

## Contexto
Trabajas como analista de datos en una tienda de electrónica y te han proporcionado un conjunto de datos con las ventas mensuales de varios productos. Tu tarea es realizar un análisis detallado utilizando diferentes técnicas de visualización.

## Conjunto de Datos
El conjunto de datos contiene las ventas mensuales (en unidades) de tres categorías de productos: teléfonos, laptops y tabletas.

```python
import matplotlib.pyplot as plt
import pandas as pd

# Creación del DataFrame con los datos de ventas
datos = {
    'Mes': ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio',
            'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre'],
    'Teléfonos': [120, 110, 130, 140, 150, 160, 155, 150, 145, 140, 135, 130],
    'Laptops': [80, 85, 90, 95, 100, 110, 105, 100, 95, 90, 85, 80],
    'Tabletas': [50, 60, 65, 70, 75, 80, 85, 90, 95, 100, 105, 110]
}
df = pd.DataFrame(datos)
df
```

```python
df['Teléfonos']
```

```python
plt.style.use('seaborn-darkgrid')
```

```python
fig, axs = plt.subplots(3, 1,  figsize=(10, 12))
```

```python
axs[0].plot(df['Mes'], df['Teléfonos'], marker='o', linestyle='-', color='blue')
axs[0].set_title('Ventas mensuales de telefonos')

axs[1].plot(df['Mes'], df['Laptops'], marker='s', linestyle='--', color='red')
axs[1].set_title('Ventas mensuales de laptops')

axs[2].plot(df['Mes'], df['Tabletas'], marker='x', linestyle=':', color='green')
axs[2].set_title('Ventas mensuales de tabletas')
plt.tight_layout()
plt.show()
```

```python
fig.show()
```

```python
ventas_totales = df.sum()[1:]
```

```python
#plt.figure(figsize=(8,8))
plt.pie(ventas_totales, labels=ventas_totales.index, autopct='%1.2f %%')
plt.title('Pie graph')
plt.show()
```

```python
plt.style.use('classic')
df.boxplot(column=['Teléfonos', 'Laptops', 'Tabletas'])
plt.title('Boxplot tarea')
plt.show()
```
